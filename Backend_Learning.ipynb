```python
#Find 2nd Largest Number without Sorting
def large(arr):
  if len(arr) < 2:
      return

  if arr[0] > arr[1]:
      large, second = arr[0], arr[1]
  else:
      large, second = arr[1], arr[0]

  for j in range(2, len(arr)):
      if arr[j] > large:
          second = large
          large = arr[j]
      elif arr[j] > second and arr[j] != large:
          second = arr[j]

  print(arr)
  print("The second Largest Number is: ", second)

n = int(input("Enter the Number of Elements: ").strip())
arr = []
for i in range(n):
  arr.append(int(input().strip()))
large(arr)
```
